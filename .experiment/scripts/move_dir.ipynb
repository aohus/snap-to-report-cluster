```python
import os
from pathlib import Path
import shutil

def archive(src, dst):
    src_dir = src if src else Path("./exp_results/archive/단풍/map")
    dst_dir = dst if dst else Path("./exp_results/archive/단풍/mosaics")
    images = [p for p in src_dir.glob('*')]

    for image in images:
        shutil.move(image, dst_dir / image.name)
```
